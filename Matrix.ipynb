```python
import numpy as np
import time
from scipy.linalg import solve
from scipy.sparse.linalg import dsolve
from scipy.sparse import csr_matrix, csc_matrix, coo_matrix, lil_matrix
```

```python
def nam_val(var):
    for (s, v) in globals().items():
        if id(v)==id(var):
            print(s,'=' )
            print(var)
            
def nam_vals(vars):
    for var in vars:
        for (s, v) in globals().items():
            if id(v)==id(var):
                print(s,'=' )
                print(v)
```

```python
##definition of matrix
list1=[[0,1,2],[3,4,5]]
mat1=np.array(list1)
nam_vals([list1,mat1])
```

```python
##Operagion of matrix
mat1=np.array([[1,2,3],[4,5,6]])
mat2=np.array([[1,1,1],[1,1,1]])
mat1T=mat1.T
mat2T=mat2.T
nam_vals([mat1,mat1T,mat2,mat2T])
matZero=np.zeros((2,3))
matOne=np.ones((2,3))
nam_vals([matZero,matOne])
print("mat1*mat2 =")
print(mat1*mat2)
print("dot(mat1T,mat2)")
print(np.dot(mat1T,mat2))
print("dot(mat1,mat2T)")
print(np.dot(mat1,mat2T))
mat3=np.array([[1,2],[4,5]])
mat3I=np.linalg.inv(mat3)
nam_vals([mat3,mat3I])
for c in mat2:
    print(c)
```

```python
# Solve Ax=b
# Definition A, b
L=1000
A=np.zeros((L,L))
for i in range(0,A.shape[0]):
    for j in range(0,A.shape[1]):
        if(i==j):
            A[i][j]=1.
        elif(np.abs(i-j)==1):
            A[i][j]=-0.5
b=np.ones((L,1))
x=np.zeros((L,1))
nam_vals([A,b,x])
```

```python
# Solve Ax=b
# Calculation invers matrix
x=np.zeros((L,1))
x=np.dot(np.linalg.inv(A),b)
nam_val(x)
print(np.dot(A,x))
```

```python
# Solve Ax=b
# Iterative method
x=np.zeros((L,1))
error=1
counter=0
while (error>1e-9 and counter<1000):
    counter+=1
    error=0
    sum=0
    for (xi,i) in zip(x,range(0,len(x))):
        old=xi
        a=A[i,0:len(x)]
        u=a[i]*xi+b[i]
        for j in range(0,len(x)):
            u-=a[j]*x[j]
        xi[:]=u/a[i]
    db=np.dot(A,x)-b
    error=np.sqrt(((db*db).sum())/((b*b).sum()))
    print("iteration=",counter,"residual=",error)    
nam_val(x)
```

スパースな行列AでのAx=bをscipyを使って解く

```python
#Use scipy.sparse (Sparse Matrix)
data = []
row = []
col = []
for i in range(0,L):
    for j in range(0,L):
        if(i==j):
            data.append(1)
            row.append(i)
            col.append(j)
        elif(np.abs(i-j)==1):
            data.append(-0.5)
            row.append(i)
            col.append(j)
nam_vals([data,row,col])
```

```python
Acsr = csr_matrix((data, (row, col)), shape=(L, L))
#Acsr = csr_matrix(A)
bcsr=csr_matrix(b)
xcsr=dsolve.spsolve(Acsr,bcsr)
nam_val(xcsr)
```

```python
def func_solve1(A,b):
    output = solve(A,b)
    return output
def func_solve2(A,b):
    Acsr = csr_matrix(A)
    bcsr = csr_matrix(b)
    output = dsolve.spsolve(Acsr,bcsr)
    return output
```

```python
st=time.time()
x = func_solve1(A,b)
et=time.time()
print(et-st)
st=time.time()
x = func_solve2(A,b)
et=time.time()
print(et-st)
```
