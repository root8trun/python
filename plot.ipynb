```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# データ生成
x = np.linspace(0, 10, 100)
y = x + np.random.randn(100) 

# プロット
plt.plot(x, y, label="test")

# 凡例の表示
plt.legend()

# プロット表示(設定の反映)
plt.show()
```

```python
# データ生成
x = np.linspace(0, 10, 100)
y = x + np.random.randn(100) 

# Figureの初期化
fig = plt.figure(figsize=(8, 4)) #...1

# Figure内にAxesを追加()
ax = fig.add_subplot(111) #...2
ax.plot(x, y, label="test") #...3

# 凡例の表示
plt.legend()

# プロット表示(設定の反映)
plt.show()
```

```python
# データ生成
x = np.linspace(0, 10, 100)
y = x + np.random.randn(100) 

# Figureの初期化
fig = plt.figure(figsize=(8, 6)) #...1

# Figure内にAxesを追加()
ax = fig.add_subplot(221) #...2
ax.plot(x, y, label="test1") #...3

ax2 = fig.add_subplot(222) #...2
ax2.plot(x*2, y, label="test2") #...3

ax3 = fig.add_subplot(223) #...2
ax3.plot(x, y*2, label="test31") #...3
ax3.plot(x, y*3, label="test32") #...3

ax4 = fig.add_subplot(224) #...2
ax4.plot(x*2, y*2, label="test4") #...3

# 凡例の表示
ax.legend(loc = 'upper right') #凡例
ax2.legend(loc = 'upper right') #凡例
ax3.legend(loc = 'upper right') #凡例
ax4.legend(loc = 'upper right') #凡例

# プロット表示(設定の反映)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = x + np.random.randn(100) 

fig, ax = plt.subplots(2,2,figsize=(6,6))

ax[0][0].plot(x, y, label="test1")
ax[0][1].plot(x*2, y, label="test1")
ax[1][0].plot(x, y*2, label="test1")
ax[1][0].plot(x, y*3, label="test1")
ax[1][1].plot(x*2, y*2, label="test1")

plt.show()
```
